```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo)
    return dict
```

```python
data_filename = '/volSSD/training_data/imagenet64/train_data_batch_1_py2_0idx'
label_filename = '/volSSD/training_data/imagenet64/map_clsloc.txt'

w = unpickle( data_filename )
df_l = pd.read_table( label_filename, sep=' ', header=None )
```
